## Pivoting tables

Pivot tables are useful for some of the problems. We will talk about them next Monday, but here is a little example from the pandas documentation (http://pandas.pydata.org/pandas-docs/stable/reshaping.html). We first create a dataframe, you **do not need to understand** this cell, just execute it and have a look at the resulting dataframe:

````python
# We first create a dataframe, you do not need to understand this function:
import pandas.util.testing as tm; tm.N = 3
import numpy as np
def unpivot(frame):
    N, K = frame.shape
    data = {'value' : frame.values.ravel('F'),
            'variable' : np.asarray(frame.columns).repeat(N),
            'date' : np.tile(np.asarray(frame.index), K)}
    return pd.DataFrame(data, columns=['date', 'variable', 'value'])
df_hint = unpivot(tm.makeTimeDataFrame())
df_hint
````

Observe that we have 12 rows of data, one for each combination of **date** and **variable** values. The data in the table would be easier to understand if we could use the values of **date** and **variable** as labels. The `pivot` method does just that:

````python
df_hint.pivot(index='date', columns='variable', values='value')
````

We could also *pivot in the other direction*, using **date** for the columns and **variable** for the rows:

````python
df_hint.pivot(index='variable', columns='date', values='value')
````

Another option is the `pivot_table` method, which combines the functionality of `groupby` and `pivot`:

````python
df_pt = pd.DataFrame({'name': list("ABCDBADC" * 2), 'major': ('java,math,' *8).split(',')[:16],
                      'scores': list(range(40, 90, 3))[:16]})
df_pt
````

````python
pd.pivot_table(df_pt, columns='major', index='name', values='scores', aggfunc=np.max)
````

Observe that in each field we now have the maximum value of the two scores. Possible values for `aggfunc` are `np.max`, `np.min`, `np.sum`, `np.mean`, and many more. We can even combine them as follows:

````python
pd.pivot_table(df_pt, columns='major', index='name', values='scores', 
               aggfunc=[np.max, np.mean])
````
